```python
import pandas as pd
import numpy as np
from sklearn.metrics import log_loss
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
#load data
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

#separate the response 
train_response = train[['toxic','severe_toxic','obscene','threat','insult','identity_hate']]

#countvectorizer
cv = CountVectorizer()
x_train = cv.fit_transform(train.comment_text.values)
x_test = cv.transform(test.comment_text.values)
```

```python
test_predict = pd.DataFrame(columns = train_response.columns)

#loop through the response and fit a classifier for each one

for col in train_response.columns:
    rf = RandomForestClassifier()
    rf.fit(x_train, train_response[col].values)
    test_predict[col] = rf.predict_proba(x_test)[:,1]
```
